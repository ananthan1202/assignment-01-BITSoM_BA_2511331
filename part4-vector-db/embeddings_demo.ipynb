```python
sentences = [

# Cricket
"Cricket is one of the most popular sports in India.",
"Virat Kohli is known for his aggressive batting style.",
"A fast bowler can change the outcome of a cricket match.",
"Test cricket requires patience and strategy.",

# Cooking
"Cooking at home is healthier than eating outside.",
"Adding spices can enhance the flavor of a dish.",
"Baking requires precise measurements of ingredients.",

# Cybersecurity
"Cybersecurity protects systems from digital attacks.",
"Strong passwords help prevent unauthorized access.",
"Phishing attacks try to steal sensitive information."
]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)

print("Embedding shape:", embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)

similarity_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm')
plt.title('Sentence Similarity Heatmap')
plt.show()
```

```python
query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])
similarities = cosine_similarity(query_embedding, embeddings)[0]
```

```python
import numpy as np
top_indices = np.argsort(similarities)[-2:][::-1]
print("Query:", query)

for i in top_indices:
  print("\nSimilar Sentences:",
sentences[i])
  print("Similarity Score:",
similarities[i])
```
